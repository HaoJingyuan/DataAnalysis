```python
from lxml import objectify
```

```python
path = 'datasets/mta_perf/Performance_MNR.xml'
paresd = objectify.parse(open(path))
root = paresd.getroot()
```

```python
data = []
skip_filed = ['PARENT_SEQ','INDICATOR_SEQ','DESIRED_CHANGE','DECIMAL_PLACES']
for elt in root.INDICATOR:
    el_data={}
    for ch in elt.getchildren():
        if ch.tag in skip_filed:
            continue
        el_data[ch.tag] = ch.pyval
    data.append(el_data)
```

```python
perf = pd.DataFrame(data)
```

```python
perf.head()
```
